```python
def get_healthcare_response(user_query):
    # Modify prompt to include symptom analysis
    prompt = """
    You are a professional healthcare assistant.
    If the user mentions a symptom, analyze it and suggest possible conditions, along with diagnosis methods.
    If they mention a disease, provide causes, symptoms, treatment, and prevention.
    Always remind users to consult a doctor for accurate diagnosis.

    User query: {user_query}
    """
    formatted_prompt = prompt.format(user_query=user_query)

    # Prepare the input as a human message
    human_message = HumanMessage(content=formatted_prompt)

    # Stream agent response
    response_chunks = []
    default_config = {"configurable": {"thread_id": "healthcare_bot"}}

    for chunk in agent.stream({"messages": [human_message]}, default_config):
        if isinstance(chunk, dict) and 'agent' in chunk:
            msgs = chunk['agent'].get('messages', [])
            if msgs and hasattr(msgs[0], 'content'):
                response_chunks.append(msgs[0].content)

    # Get final response
    final_result = response_chunks[-1] if response_chunks else "I couldn't find an answer. Please consult a healthcare professional."
    return final_result.strip()
```

```python
# List of symptom queries
symptom_queries = [
    "I have a high fever and joint pain, what could it be?",
    "I feel chest pain and shortness of breath, should I be worried?",
    "I'm experiencing severe headaches and nausea, what should I do?",
    "My throat is sore and I have difficulty swallowing, is it serious?",
    "I have red rashes and a high fever, what could this be?"
]

# Loop through each symptom and get chatbot response
for query in symptom_queries:
    response = get_healthcare_response(query)
    print(f"\n🩺 **User Query:** {query}")
    print(f"🤖 **Chatbot Response:** {response}\n")
```
